```python
from requests_html import HTMLSession

session = HTMLSession()

response = session.get('http://thequotesmaster.com/2019/01/350popular-list-of-slogans-taglines-of-top-brands/')

img_a = response.html.find("li")
links = []
for link in img_a:
    links.append(link.text)

links = links[11:-8]
links
```

```python
links[0].split('–')[-1].strip(' ')
```

```python
for link in range(len(links)):
    l = links[link].split('–')[-1].strip(' ')
    print(l)
```

```python
#clean the text file
file = open('adLines.txt','w')
file.write('')
file.close()
```

```python
#append the content
file = open('adLines.txt','a')
for link in range(len(links)):
    l = links[link].split('–')[-1].strip(' ')
    file.write('\n' + l)
file.close()
```
