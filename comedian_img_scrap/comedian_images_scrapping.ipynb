```python
from requests_html import HTMLSession

session = HTMLSession()

# we use session.get to get the contents of the url
response = session.get('https://scrapsfromtheloft.com/comedy/')

img_a = response.html.find("div.fusion-image-wrapper a")
links = []
for link in img_a:
    links.append(link.attrs.get('href'))

img_a
```

```python
import pandas as pd

df = pd.DataFrame(columns=['id', 'images', 'title', 'scripts'])
for link in links:
    response = session.get(link)
    
    try: 
    
        page_id = link.split('/')[-1]

        images = [img.attrs.get('src') for img in response.html.find('.attachment-full')]

        title = response.html.find('.entry-title', first=True).text
        
        script = []
        
        #going into individual link to grab table data
        for  i in range(1,(len(response.html.find('p'))-2)):
            p = response.html.find('p')[i].text
            script.append(p)
            #print(str(i) + " of " + str(len(response.html.find('p'))))
        print("Processed no." + str(links.index(link)) + " transcript")
            
        df = df.append({'id': page_id,
            'images': images,
            'title': title,
            'scripts': script}, ignore_index=True)

    except Exception as e:
        # Print the error
        print(e)
        # and print the associated link
        print('failed on ' + link)
```
